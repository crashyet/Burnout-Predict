## 1. Analisis Permasalahan & Solusi Utama

### 1.1 Identifikasi Permasalahan

Kesehatan mental pekerja di Indonesia saat ini berada dalam kondisi yang mengkhawatirkan. Berdasarkan berbagai sumber yang telah diverifikasi, ditemukan **5 permasalahan utama** yang menjadi latar belakang proyek ini.

**1. Burnout Sudah Menjadi Krisis Nasional**
Data riset dari Fakultas Kedokteran Universitas Indonesia (FKUI) menunjukkan bahwa **83% tenaga kesehatan** di Indonesia mengalami gejala burnout tingkat sedang hingga berat selama pandemi [1]. Hal ini mengindikasikan bahwa burnout merupakan isu serius yang meluas ke berbagai sektor pekerjaan dan perlu perhatian nasional.

**2. Burnout Tidak Terdeteksi Sampai Terlambat**
Kebanyakan individu baru menyadari kondisi burnout setelah mencapai tahap kronis — ditandai dengan penurunan produktivitas drastis, memburuknya hubungan interpersonal, atau transisi ke fase depresi klinis. Saat ini, belum tersedia sistem **peringatan dini** yang mampu memberikan sinyal sebelum kondisi memburuk secara signifikan.

**3. Tidak Ada Alat Pemantau Emosi Harian yang Terintegrasi**
Aplikasi jurnal harian, pemantau tidur, dan manajemen produktivitas saat ini beroperasi secara terpisah (terfragmentasi). Tidak adanya platform tunggal yang menghubungkan variabel-variabel ini menyebabkan pengguna sulit mendapatkan gambaran kondisi mental yang utuh dan komprehensif.

**4. Stigma dan Keterbatasan Akses ke Profesional**
Menurut laporan WHO (2022), di negara berpenghasilan rendah dan menengah, diperkirakan **lebih dari 75%** orang dengan gangguan mental tidak menerima pengobatan sama sekali [2]. Di Indonesia, stigma sosial yang kuat menjadi penghalang utama bagi pekerja untuk mencari bantuan profesional secara terbuka.

**5. Fragmentasi Data: Emosi dan Aktivitas Fisik Belum Terintegrasi**
Riset dalam bidang *Affective Computing* membuktikan bahwa pendekatan multimodal — menggabungkan teks dan data perilaku — memberikan akurasi prediksi kondisi mental yang jauh lebih tinggi [3]. Namun, data kualitatif (teks jurnal) dan kuantitatif (jam tidur & kerja) selama ini tersimpan terpisah dan tidak pernah dianalisis bersamaan.

##### Load English Emotion Dataset

```python
LABEL_MAP_EN = {'joy': 'happy', 'sadness': 'sadness',
                'anger': 'anger', 'fear': 'fear', 'love': 'love'}

frames = []
with zipfile.ZipFile(PATH_ENGLISH_ZIP, 'r') as zf:
    for split in ['train.txt', 'val.txt', 'test.txt']:
        with zf.open(split) as f:
            tmp = pd.read_csv(f, sep=';', header=None,
                              names=['text', 'label'], encoding='utf-8')
            tmp['split'] = split.replace('.txt', '')
            frames.append(tmp)

df_english = pd.concat(frames, ignore_index=True)
df_english['label'] = df_english['label'].str.strip().map(LABEL_MAP_EN)

n_before   = len(df_english)
df_english = df_english.dropna(subset=['label']).reset_index(drop=True)
n_dropped  = n_before - len(df_english)

print(f'   Baris total     : {n_before:,}')
print(f'   Surprise dihapus: {n_dropped:,}')
print(f'   Shape akhir     : {df_english.shape}')
print(f'   Labels          : {sorted(df_english["label"].unique().tolist())}')
display(df_english['split'].value_counts().rename('jumlah').to_frame())
display(df_english.head(5))
```

Dataset bahasa Inggris diekstrak langsung dari dalam `archive.zip` tanpa perlu mengekstrak file terlebih dahulu, lalu ketiga split digabung sekaligus. Label dipetakan ke skema 5-emosi proyek via `LABEL_MAP_EN` — label yang tidak terdaftar otomatis menjadi `NaN` dan langsung di-*drop*.

Hasilnya **719 baris `surprise` tereliminasi**, menyisakan **19.281 baris** dengan 5 label yang seragam. Dataset ini jauh lebih besar dari sumber Indonesia lainnya (15.428 train saja sudah melebihi total Twitter + Ricco48 + IndoNLU). > **Langkah selanjutnya:** Seluruh teks perlu diterjemahkan ke Bahasa Indonesia sebelum bisa digabung dengan dataset lain — proses ini dilakukan pada tahap *Cleaning* menggunakan `GoogleTranslator` dengan mekanisme *checkpoint* untuk menghindari kehilangan progres jika sesi terputus.

##### Fungsi Cleaning Teks

```python
def clean_text(text, kamus=None):
    """Membersihkan teks: lowercase, normalisasi singkatan, hapus noise."""
    text = str(text).lower()
    if kamus:
        text = ' '.join(kamus.get(w, w) for w in text.split())
    text = re.sub(r'http\S+|www\S+',  '', text)
    text = re.sub(r'@\w+|#\w+',       '', text)
    text = re.sub(r'\d+',             '', text)
    text = re.sub(r'[^a-zA-Z\s]',     '', text)
    text = re.sub(r'\s+',             ' ', text).strip()
    return text

print('✅ Fungsi clean_text siap digunakan.')
```

Mendefinisikan fungsi pembersih teks yang akan digunakan seragam di semua dataset NLP. Pipeline pembersihan berjalan dalam urutan berikut:

1. *Lowercase* seluruh teks
2. Normalisasi singkatan via `KAMUS_DICT` (opsional — dilewati jika `kamus=None`)
3. Hapus URL dan domain web
4. Hapus mention (`@user`) dan hashtag (`#topik`)
5. Hapus angka
6. Hapus semua karakter selain huruf dan spasi
7. Rapikan spasi ganda

> **Langkah selanjutnya:** Fungsi ini dipanggil dengan `clean_text(teks, KAMUS_DICT)` untuk dataset Indonesia, dan `clean_text(teks)` tanpa kamus untuk teks yang sudah diterjemahkan — karena hasil terjemahan sudah dalam bentuk kata baku.

```python
from deep_translator import GoogleTranslator

PATH_TRANSLATED = os.path.join(DATASET_DIR, 'english_emotion_translated.csv')
PATH_CHECKPOINT = os.path.join(DATASET_DIR, 'english_translate_checkpoint.csv')

def translate_one(translator_obj, text, max_len=4500):
    text = str(text).strip()[:max_len]
    if not text:
        return text
    try:
        result = translator_obj.translate(text)
        return result if result else text
    except Exception:
        return text

if os.path.exists(PATH_TRANSLATED):
    df_en_translated = pd.read_csv(PATH_TRANSLATED)
    print(f'✅ Cache ditemukan — Shape: {df_en_translated.shape}')
else:
    texts_all = df_english['text'].astype(str).tolist()
    total     = len(texts_all)

    if os.path.exists(PATH_CHECKPOINT):
        df_ckpt   = pd.read_csv(PATH_CHECKPOINT)
        start_idx = len(df_ckpt)
        results   = df_ckpt['tweet_clean'].tolist()
        print(f'Checkpoint ditemukan — lanjut dari {start_idx:,}/{total:,}')
    else:
        start_idx, results = 0, []
        print(f'Mulai translate {total:,} baris...')

    translator = GoogleTranslator(source='en', target='id')

    for idx in range(start_idx, total):
        for attempt in range(5):
            try:
                results.append(translate_one(translator, texts_all[idx]))
                break
            except Exception:
                if attempt == 4:
                    results.append(texts_all[idx])
                else:
                    time.sleep(3.0 * (attempt + 1))
        time.sleep(0.1)

        if (idx + 1) % 500 == 0 or (idx + 1) == total:
            ckpt = df_english.iloc[:len(results)][['text', 'label']].copy()
            ckpt['tweet_clean'] = results
            ckpt.to_csv(PATH_CHECKPOINT, index=False)
            print(f'   Checkpoint: {idx+1:,}/{total:,} ({(idx+1)/total*100:.1f}%)')

    df_en_translated = df_english[['text', 'label']].copy()
    df_en_translated['tweet_clean'] = results
    df_en_translated.to_csv(PATH_TRANSLATED, index=False)

    if os.path.exists(PATH_CHECKPOINT):
        os.remove(PATH_CHECKPOINT)
    print(f'✅ Translate selesai!')

display(df_en_translated[['text', 'tweet_clean', 'label']].head(5))
```

Proses terjemahan 19.281 teks ke Bahasa Indonesia menggunakan `GoogleTranslator` dengan dua mekanisme keamanan:
- **Retry otomatis** hingga 5 kali dengan jeda bertahap jika terjadi error koneksi.
- **Checkpoint setiap 500 baris** — jika sesi Colab terputus, proses dapat dilanjutkan dari baris terakhir tanpa mengulang dari awal.

Karena file cache sudah tersedia, sel ini langsung membaca hasil terjemahan tanpa menjalankan ulang proses translate. Dari sampel output, terjemahan berjalan dengan baik — makna emosi tetap terjaga (`"hopeless"` → `"putus asa"`, `"grouchy"` → `"kesal"`).

## 6. Synthetic Data Generation

### Mengapa Synthetic Data Diperlukan?

Dataset burnout bersifat *cross-sectional* — setiap baris hanya merepresentasikan satu *snapshot* kondisi seorang pekerja pada satu titik waktu, **bukan rekaman harian**. Sementara itu, model LSTM membutuhkan data sekuensial (*time-series*) yang merekam perubahan kondisi dari hari ke hari.

Setelah pencarian menyeluruh di sumber publik, tidak ditemukan dataset yang memiliki kombinasi `work_hours` harian, `sleep_hours` harian, dan `burnout_score` harian dalam satu tabel. Oleh karena itu, **dataset time-series dibangun sendiri** menggunakan dataset burnout asli sebagai *anchor* (titik awal simulasi per user).

> ⚠️ **Catatan Penting:** Data synthetic ini **bukan data nyata** — dibuat berdasarkan statistik dan pola dari data asli. Model LSTM yang dilatih dengan data ini harus divalidasi secara hati-hati di data nyata, karena distribusi kelas `High` yang jauh lebih besar dari kondisi riil berpotensi membuat model terlalu sensitif dalam memprediksi burnout berat.

#### Statistik Anchor per Burnout Level

```python
print('=== STATISTIK ANCHOR PER BURNOUT LEVEL ===\n')
print('Fitur rata-rata per Level:')
display(
    df_burnout_clean.groupby('burnout_level')
    [['total_work_hours_per_week', 'sleep_hours', 'burnout_score']]
    .mean().round(2)
)
print('\nburnout_score range per Level:')
display(
    df_burnout_clean.groupby('burnout_level')['burnout_score']
    .describe().round(2)
)
```

Statistik ini dihitung dari data asli untuk dijadikan **dasar logika simulasi** — memastikan data synthetic yang dihasilkan mencerminkan pola nyata, bukan angka acak.

Pola antar level sangat jelas dan monoton:

| Level | Jam Kerja/Minggu | Jam Tidur | Burnout Score |
| :--- | ---: | ---: | ---: |
| `Low` | 52,06 | 6,54 | 1,82 |
| `Moderate` | 60,84 | 6,22 | 4,25 |
| `High` | 69,12 | 5,66 | 6,89 |

Semakin tinggi level burnout, jam kerja naik dan jam tidur turun secara konsisten. Dari kolom `burnout_score range`, batas antar level juga terlihat jelas dan tidak tumpang tindih — `Low` maksimal 3,5, `Moderate` mulai 3,6 hingga 6,5, dan `High` mulai 6,6 ke atas. Batas-batas ini yang kemudian digunakan sebagai threshold pada loop simulasi.

> **Langkah selanjutnya:** Threshold simulasi diturunkan langsung dari statistik ini — `THRESH_BURNOUT_MOD = 3.6`, `THRESH_BURNOUT_HI = 6.6`, `THRESH_WORK_HEAVY = 69.12/5 ≈ 12 jam/hari`, dan `THRESH_SLEEP_HIGH = 5.66 jam` — sebelum proses *stratified sampling anchor* dimulai.

#### Stratified Sampling Anchor

```python
SEED   = 42
N_DAYS = 30
WINDOW = 7
np.random.seed(SEED)

df_high     = df_burnout_clean[df_burnout_clean['burnout_level'] == 'High']
df_moderate = df_burnout_clean[df_burnout_clean['burnout_level'] == 'Moderate'].sample(500, random_state=SEED)
df_low      = df_burnout_clean[df_burnout_clean['burnout_level'] == 'Low'].sample(500, random_state=SEED)
anchors     = pd.concat([df_high, df_moderate, df_low]).reset_index(drop=True)

print(f'Anchor users: High={len(df_high)} | Moderate=500 | Low=500 | Total={len(anchors)}')
print(f'Estimasi baris harian : {len(anchors) * N_DAYS:,}')
print(f'Estimasi sampel LSTM  : {len(anchors) * (N_DAYS - WINDOW):,}')
```

Memilih titik awal simulasi dari data asli dengan komposisi yang disengaja tidak seimbang: semua **65 data `High`** diambil seluruhnya karena merupakan kelas minoritas yang paling kritis, sementara `Moderate` dan `Low` masing-masing diambil **500 sampel** acak agar simulasi tidak terlalu didominasi kondisi normal.

Dari 1.065 anchor user yang disimulasikan selama 30 hari, estimasi output yang dihasilkan:
- **31.950 baris** data harian mentah
- **24.495 sampel** LSTM setelah sliding window 7→1 diterapkan (setiap user kehilangan 7 hari pertama sebagai window awal)

> **Langkah selanjutnya:** Setiap anchor user disimulasikan hari per hari dalam loop menggunakan logika korelasi yang diturunkan dari statistik — kerja berlebih memicu penurunan tidur, akumulasi lembur mendorong kenaikan burnout score secara progresif.

#### Synthetic Data Generation (Loop Simulasi)

```python
THRESH_WORK_SAFE   = 10.4
THRESH_WORK_HEAVY  = 12.0
THRESH_SLEEP_HIGH  = 5.66
THRESH_BURNOUT_MOD = 3.6
THRESH_BURNOUT_HI  = 6.6

all_daily = []

for uid, row in anchors.iterrows():
    work_h   = row['total_work_hours_per_week'] / 5
    sleep_h  = row['sleep_hours']
    burnout  = row['burnout_score']
    consec   = 0

    for day in range(N_DAYS):
        work_h  = np.clip(work_h  + np.random.uniform(-1.5, 1.5), 4, 19)
        sleep_h = np.clip(sleep_h + np.random.uniform(-0.5, 0.5), 3, 10)

        if work_h > THRESH_WORK_HEAVY:
            sleep_h = np.clip(sleep_h - np.random.uniform(0.1, 0.4), 3, 10)
            consec += 1
        else:
            consec = max(0, consec - 1)

        if consec >= 3:
            delta = np.random.uniform(0.2, 0.5)
        elif work_h > THRESH_WORK_HEAVY or sleep_h < THRESH_SLEEP_HIGH:
            delta = np.random.uniform(0.1, 0.3)
        elif work_h < THRESH_WORK_SAFE and sleep_h > 6.54:
            delta = np.random.uniform(-0.2, -0.05)
        else:
            delta = np.random.uniform(-0.1, 0.1)

        burnout = np.clip(burnout + delta, 1.0, 10.0)
        level   = ('High'     if burnout >= THRESH_BURNOUT_HI
                   else 'Moderate' if burnout >= THRESH_BURNOUT_MOD
                   else 'Low')

        all_daily.append({
            'user_id'              : f'user_{uid:04d}',
            'day'                  : day + 1,
            'work_hours_per_day'   : round(work_h, 2),
            'sleep_hours'          : round(sleep_h, 2),
            'burnout_score'        : round(burnout, 2),
            'burnout_level'        : level,
            'consecutive_overwork' : consec,
            'anchor_level'         : row['burnout_level'],
        })

df_synthetic = pd.DataFrame(all_daily)
print(f'✅ Simulasi selesai — Shape: {df_synthetic.shape}')
display(df_synthetic['burnout_level'].value_counts())
display(df_synthetic.head(5))
```

Setiap anchor user disimulasikan selama 30 hari dengan logika korelasi berbasis domain:
- Kerja **> 12 jam/hari** → tidur berkurang + `consecutive_overwork` naik
- Lembur **≥ 3 hari berturut-turut** → burnout naik **+0,2–0,5/hari** (dampak kumulatif terkuat)
- Kerja **< 10,4 jam** DAN tidur **> 6,54 jam** → burnout turun (kondisi recovery)
- Semua nilai dijaga dalam batas wajar dengan `np.clip`

Dari sampel `user_0000`, pola simulasi bekerja sesuai logika — kerja konsisten >12 jam/hari menyebabkan tidur turun progresif (6,03 → 5,02 jam) dan `consecutive_overwork` terus naik, mendorong burnout meningkat dari 6,82 ke 8,13 dalam 5 hari pertama.

Distribusi hasil simulasi jauh lebih seimbang dibanding data asli — `High` 40,0%, `Moderate` 29,3%, `Low` 30,7% — karena delta burnout yang agresif secara progresif mendorong banyak user masuk ke zona `High` setelah lembur berturut-turut.

> **Langkah selanjutnya:** Validasi korelasi dilakukan untuk memastikan arah hubungan antar variabel pada data synthetic konsisten dengan kondisi nyata sebelum data ini digunakan lebih lanjut.

#### Validasi Korelasi Synthetic

```python
r_syn_work,  _ = stats.pearsonr(df_synthetic['work_hours_per_day'], df_synthetic['burnout_score'])
r_syn_sleep, _ = stats.pearsonr(df_synthetic['sleep_hours'],        df_synthetic['burnout_score'])

print(f'Korelasi work_hours  vs burnout: {r_syn_work:+.3f}  '
      f'{"✅ POSITIF" if r_syn_work > 0 else "⚠️ NEGATIF"}')
print(f'Korelasi sleep_hours vs burnout: {r_syn_sleep:+.3f}  '
      f'{"✅ NEGATIF" if r_syn_sleep < 0 else "⚠️ POSITIF"}')

print('\nRata-rata per burnout_level:')
display(
    df_synthetic.groupby('burnout_level')
    [['work_hours_per_day', 'sleep_hours', 'burnout_score']]
    .mean().round(2)
)
```

Dua kondisi wajib diverifikasi untuk memastikan data synthetic logis secara domain:
- `work_hours` harus berkorelasi **positif** dengan burnout → terpenuhi (+0,531)
- `sleep_hours` harus berkorelasi **negatif** dengan burnout → terpenuhi (−0,763)

Korelasi `sleep_hours` (−0,763) lebih kuat dibanding `work_hours` (+0,531) — menunjukkan bahwa dalam simulasi ini, kualitas tidur adalah prediktor burnout yang lebih sensitif dibanding jam kerja semata.

Rata-rata per level juga menunjukkan pola yang monoton dan masuk akal: semakin tinggi level burnout, jam kerja naik dan jam tidur turun secara konsisten. Grup `High` rata-rata hanya tidur **3,75 jam/malam** — mendekati batas minimum 3 jam yang ditetapkan `np.clip`, mencerminkan kondisi ekstrem yang realistis untuk burnout berat.

> **Langkah selanjutnya:** Sanity check nilai tidak wajar dilakukan untuk memastikan tidak ada nilai di luar batas yang ditetapkan `np.clip` akibat akumulasi noise selama 30 hari simulasi.

#### Validasi Nilai Tidak Wajar

```python
checks = {
    'work_hours < 4'   : (df_synthetic['work_hours_per_day'] < 4).sum(),
    'work_hours > 19'  : (df_synthetic['work_hours_per_day'] > 19).sum(),
    'sleep_hours < 3'  : (df_synthetic['sleep_hours'] < 3).sum(),
    'sleep_hours > 10' : (df_synthetic['sleep_hours'] > 10).sum(),
    'burnout < 1'      : (df_synthetic['burnout_score'] < 1).sum(),
    'burnout > 10'     : (df_synthetic['burnout_score'] > 10).sum(),
    'NaN total'        : df_synthetic.isnull().sum().sum(),
}
for check, count in checks.items():
    print(f'  {"✅" if count == 0 else "⚠️"} {check:<25}: {count}')
```

Memeriksa 7 kondisi batas untuk memastikan tidak ada nilai anomali yang lolos dari `np.clip` selama 30 hari akumulasi noise. Semua kondisi mengembalikan **0 pelanggaran** — tidak ada jam kerja di bawah 4 atau di atas 19, tidak ada jam tidur di bawah 3 atau di atas 10, tidak ada burnout di luar skala 1–10, dan tidak ada NaN.

> **Langkah selanjutnya:** Data synthetic dinyatakan valid dan siap dikonversi ke format sekuensial menggunakan **Sliding Window** — setiap 7 hari data harian dijadikan satu sampel input LSTM untuk memprediksi kondisi hari ke-8.

#### Sliding Window (Format Input LSTM)

```python
FEATURE_COLS = ['work_hours_per_day', 'sleep_hours']
X_list, y_score_list, y_level_list, meta_list = [], [], [], []

for uid in df_synthetic['user_id'].unique():
    user_df = (df_synthetic[df_synthetic['user_id'] == uid]
               .sort_values('day').reset_index(drop=True))
    for i in range(len(user_df) - WINDOW):
        X_list.append(user_df.iloc[i:i+WINDOW][FEATURE_COLS].values.tolist())
        y_score_list.append(user_df.iloc[i+WINDOW]['burnout_score'])
        y_level_list.append(user_df.iloc[i+WINDOW]['burnout_level'])
        meta_list.append({'user_id': uid, 'predict_day': i+WINDOW+1,
                          'anchor_level': user_df.iloc[i+WINDOW]['anchor_level']})

X_arr       = np.array(X_list)
y_score_arr = np.array(y_score_list)
y_level_arr = np.array(y_level_list)

print(f'✅ Sliding window selesai')
print(f'   X shape   : {X_arr.shape}  → (sampel, {WINDOW} hari, {len(FEATURE_COLS)} fitur)')
print(f'   y_score   : {y_score_arr.shape}')
print(f'   y_level   : {y_level_arr.shape}')

level_counts = collections.Counter(y_level_arr)
print('\nDistribusi y_level:')
for lvl in ['Low', 'Moderate', 'High']:
    n = level_counts[lvl]
    print(f'  {lvl:<10}: {n:,} ({n/len(y_level_arr)*100:.1f}%)')
```

Mengonversi data harian menjadi format sekuensial yang dibutuhkan LSTM: setiap **7 hari berturutan** dijadikan satu sampel input (`X`) untuk memprediksi kondisi **hari ke-8** (`y`). Hasilnya `X_arr` berbentuk **(24.495, 7, 2)** — 24.495 sampel, masing-masing berisi 7 hari data dengan 2 fitur (`work_hours_per_day` dan `sleep_hours`).

Proporsi `High` meningkat dari 40,0% (data harian) menjadi **49,6%** di sampel LSTM — wajar karena user yang sudah masuk zona `High` cenderung bertahan di sana lebih lama, sehingga lebih banyak window berakhir dengan prediksi `High`.

Dalam implementasi aplikasi, prediksi burnout dihasilkan **setiap hari** sejak hari pertama pengguna mencatat data. Namun sistem akan menampilkan peringatan bahwa **akurasi prediksi meningkat seiring bertambahnya histori** — prediksi yang didukung data 30 hari atau lebih jauh lebih andal dibanding prediksi di minggu-minggu awal penggunaan.

Perlu dicatat bahwa `X_arr` ini adalah versi **baseline dengan 2 fitur mentah** dan belum final — setelah Feature Engineering di Section 8, sliding window akan dibangun ulang dengan 8 fitur sehingga shape berubah menjadi **(24.495, 7, 8)**.

#### Korelasi Jam Tidur vs Burnout Score

```python
r_sleep, p_sleep = stats.pearsonr(df_burnout_clean['sleep_hours'],
                                   df_burnout_clean['burnout_score'])
threshold_tidur  = (df_burnout_clean[df_burnout_clean['burnout_level']
                    .isin(['Moderate', 'High'])]['sleep_hours'].mean())

print(f'Pearson r  : {r_sleep:.4f}')
print(f'P-value    : {p_sleep:.4e}  {"✅ Signifikan" if p_sleep < 0.05 else "⚠️ Tidak Signifikan"}')
print(f'\n📌 Threshold peringatan dini: tidur ≤ {threshold_tidur:.1f} jam/malam')
```

Uji korelasi Pearson menunjukkan hubungan **negatif signifikan** antara jam tidur dan burnout score (r = −0,1331, p ≈ 0,000) — semakin sedikit tidur, semakin tinggi burnout. Namun kekuatan korelasinya tergolong **lemah secara praktis**.

Nilai p ≈ 0 bukan berarti hubungannya kuat — dengan 150.000 sampel, uji statistik memiliki *statistical power* yang sangat tinggi sehingga korelasi sekecil apapun akan terdeteksi signifikan. Artinya `sleep_hours` saja tidak cukup sebagai prediktor tunggal burnout, namun tetap relevan sebagai fitur pendukung.

Threshold **≤ 6,2 jam/malam** diturunkan dari rata-rata jam tidur gabungan grup `Moderate` dan `High` — bukan nilai arbitrer — dan akan digunakan sebagai batas peringatan dini pada sistem BurnoutLens.

> **Langkah selanjutnya:** Hubungan ini divisualisasikan dengan scatter plot + garis tren dan boxplot per kategori tidur pada sel berikutnya.

```python
SLEEP_ORDER = ['<5 jam', '5–6 jam', '6–7 jam', '7–8 jam', '>8 jam']

fig, axes = plt.subplots(1, 2, figsize=(15, 5.5))
fig.suptitle('BQ-2: Hubungan Jam Tidur dan Burnout Score',
             fontsize=14, fontweight='bold')

sample_bq2 = df_burnout_clean.sample(min(5000, len(df_burnout_clean)), random_state=42)

# ── Panel 1: Scatter ─────────────────────────────────────────────────────────
axes[0].scatter(sample_bq2['sleep_hours'], sample_bq2['burnout_score'],
                alpha=0.15, s=8, color='steelblue', label='Data Poin (Sampel)')

m, b = np.polyfit(df_burnout_clean['sleep_hours'],
                  df_burnout_clean['burnout_score'], 1)
x_line = np.linspace(df_burnout_clean['sleep_hours'].min(),
                     df_burnout_clean['sleep_hours'].max(), 100)

axes[0].plot(x_line, m*x_line + b, color='crimson', linewidth=2.5,
             label=f'Garis Tren (r={r_sleep:.3f})')
axes[0].axvline(threshold_tidur, color='orange', linestyle='--',
                linewidth=1.5, label=f'Threshold: {threshold_tidur:.1f} jam')
axes[0].set_title('Scatter: Sleep Hours vs Burnout Score', fontsize=12, pad=10)
axes[0].set_xlabel('Jam Tidur per Malam')
axes[0].set_ylabel('Burnout Score')
axes[0].grid(True, linestyle='--', alpha=0.3)
axes[0].legend(loc='upper right')

# ── Panel 2: Boxplot ─────────────────────────────────────────────────────────
df_bq2 = df_burnout_clean.copy()
df_bq2['sleep_bin'] = pd.cut(
    df_bq2['sleep_hours'],
    bins=[0, 5, 6, 7, 8, df_bq2['sleep_hours'].max() + 1],
    labels=SLEEP_ORDER
)

sns.boxplot(data=df_bq2, x='sleep_bin', y='burnout_score',
            order=SLEEP_ORDER,              # urutan eksplisit
            ax=axes[1], palette='Reds_r', linewidth=1.2)

# Label median di atas setiap box
medians = df_bq2.groupby('sleep_bin', observed=True)['burnout_score'].median()
for i, cat in enumerate(SLEEP_ORDER):
    if cat in medians.index:
        axes[1].text(i, medians[cat] + 0.1, f'Md: {medians[cat]:.2f}',
                     ha='center', fontsize=8, fontweight='bold')

axes[1].set_title('Distribusi Burnout Score per Kategori Tidur',
                  fontsize=12, pad=10)
axes[1].set_xlabel('Kategori Jam Tidur')
axes[1].set_ylabel('Burnout Score')
axes[1].grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.savefig('bq2_korelasi_tidur_burnout.png', dpi=150, bbox_inches='tight')
plt.show()
```

Scatter plot (kiri) memperlihatkan tren negatif yang nyata namun lemah — garis merah menurun dari kiri ke kanan, namun titik data tersebar sangat lebar di sepanjang sumbu Y. Garis oranye (threshold 6,2 jam) membagi zona berisiko dan aman secara visual.

Boxplot (kanan) memperjelas polanya per kategori: median burnout turun secara konsisten seiring bertambahnya jam tidur — dari **2,20** pada kelompok `<5 jam` hingga **1,50** pada kelompok `>8 jam`. Meski penurunannya bertahap dan tidak dramatis, arahnya konsisten tanpa pengecualian di semua kategori.

Yang perlu diperhatikan adalah distribusi tiap box sangat lebar dan banyak outlier di atas — menunjukkan bahwa tidur cukup saja tidak menjamin burnout rendah. Faktor lain seperti jam kerja kemungkinan besar turut berperan, yang akan diuji pada tahap berikutnya.

#### Fitur Dominan terhadap Burnout

```python
r_work, p_work = stats.pearsonr(df_burnout_clean['total_work_hours_per_week'],
                                  df_burnout_clean['burnout_score'])

display(pd.DataFrame({
    'Fitur'        : ['total_work_hours_per_week', 'sleep_hours'],
    'Korelasi (r)' : [round(r_work,4), round(r_sleep,4)],
    '|Korelasi|'   : [round(abs(r_work),4), round(abs(r_sleep),4)],
    'P-value'      : [f'{p_work:.2e}', f'{p_sleep:.2e}'],
    'Arah'         : ['Positif ↑', 'Negatif ↓'],
}))
fitur_dominan = 'total_work_hours_per_week' if abs(r_work) > abs(r_sleep) else 'sleep_hours'
print(f'\n📌 Fitur dominan: {fitur_dominan}')
```

Temuan ini menegaskan bahwa **beban jam kerja adalah prediktor burnout yang jauh lebih kuat** dibanding kualitas tidur secara mandiri. Namun keduanya tetap dipertahankan sebagai fitur — kombinasi keduanya (`dual_risk_flag`) terbukti memberikan sinyal risiko yang lebih kuat dari masing-masing fitur secara terpisah.

## 8. Feature Engineering


#### Import & Threshold

```python
from sklearn.preprocessing import MinMaxScaler
import joblib

THRESH_WORK_FE  = 10.4   # threshold jam kerja/hari (dari EDA BQ-3)
THRESH_SLEEP_FE = 6.2    # threshold jam tidur/malam (dari EDA BQ-2)
```

Memuat `MinMaxScaler` untuk scaling fitur dan `joblib` untuk menyimpan scaler agar dapat di-*load* ulang saat inferensi tanpa perlu fit ulang.

Dua threshold didefinisikan langsung dari temuan EDA:
- `THRESH_WORK_FE = 10.4 jam/hari` — diturunkan dari BQ-3: rata-rata jam kerja grup `Low` (52,06 jam/minggu ÷ 5)
- `THRESH_SLEEP_FE = 6.2 jam/malam` — diturunkan dari BQ-4: rata-rata jam tidur gabungan grup `Moderate` dan `High`

Keduanya akan digunakan sebagai batas pembuatan fitur biner (`overwork_flag`, `sleep_risk_flag`) pada sel berikutnya.

#### Feature Engineering Dataset Burnout

```python
df_fe = df_burnout_clean.copy()
print(f'Shape awal: {df_fe.shape}')

# -- Fitur turunan numerik --
df_fe['work_hours_per_day'] = (df_fe['total_work_hours_per_week'] / 5).round(2)

rest_temp = (24 - df_fe['work_hours_per_day'] - df_fe['sleep_hours']).clip(lower=0.1)
df_fe['work_rest_ratio'] = (df_fe['work_hours_per_day'] / rest_temp).round(4)

# [Penanganan Outlier] work_rest_ratio bisa ~180 ketika rest_temp di-clip ke 0.1
# Capping di p99 selaras dengan metode IQR yang sudah dipakai di Section 5.2
cap_99_fe = df_fe['work_rest_ratio'].quantile(0.99)
df_fe['work_rest_ratio'] = df_fe['work_rest_ratio'].clip(upper=cap_99_fe)
print(f'   ✅ work_rest_ratio di-cap di p99 = {cap_99_fe:.4f}')

df_fe['sleep_deficit'] = (8 - df_fe['sleep_hours']).round(2)   # deficit vs standar WHO

# [Encoding Variabel ke Numerik] encoding biner berbasis threshold dari EDA
df_fe['overwork_flag']   = (df_fe['work_hours_per_day'] > THRESH_WORK_FE).astype(int)
df_fe['sleep_risk_flag'] = (df_fe['sleep_hours'] <= THRESH_SLEEP_FE).astype(int)
df_fe['dual_risk_flag']  = ((df_fe['overwork_flag'] == 1) &
                             (df_fe['sleep_risk_flag'] == 1)).astype(int)

FITUR_BARU_FE = ['work_hours_per_day', 'work_rest_ratio', 'sleep_deficit',
                  'overwork_flag', 'sleep_risk_flag', 'dual_risk_flag']

print(f'✅ {len(FITUR_BARU_FE)} fitur baru dibuat:')
for f in FITUR_BARU_FE:
    print(f'   - {f}')
print(f'\nShape sesudah: {df_fe.shape}')
display(df_fe.head())
```

Enam fitur baru dibuat dari kombinasi kolom yang ada, terbagi menjadi tiga jenis:

**Fitur numerik turunan:**
- `work_hours_per_day` — konversi jam kerja mingguan ke harian (÷5) agar sebanding dengan fitur harian lainnya
- `work_rest_ratio` — rasio jam kerja terhadap jam istirahat harian (`work ÷ (24 - work - sleep)`); di-*cap* di p99 (7,47) karena bisa mencapai nilai ekstrem (~180) ketika jam istirahat mendekati nol
- `sleep_deficit` — jarak jam tidur aktual dari standar WHO 8 jam; nilai positif berarti kurang tidur

**Fitur biner berbasis threshold EDA:**
- `overwork_flag` — 1 jika kerja > 10,4 jam/hari (threshold BQ-3)
- `sleep_risk_flag` — 1 jika tidur ≤ 6,2 jam/malam (threshold BQ-4)
- `dual_risk_flag` — 1 jika **kedua** flag aktif sekaligus; sinyal kombinasi terkuat

Dari sampel output, logika flag bekerja benar — baris 0 (`9 jam kerja, 5,9 jam tidur`) hanya memicu `sleep_risk_flag`, baris 1 (`12 jam kerja, 7,1 jam tidur`) hanya memicu `overwork_flag`, dan baris 3 (`13,2 jam kerja, 6,9 jam tidur`) memicu `overwork_flag` saja karena tidurnya masih di atas threshold 6,2 jam.

> **Langkah selanjutnya:** Statistik dan distribusi flag per `burnout_level` dihitung untuk memverifikasi bahwa fitur baru ini benar-benar membedakan antar level sebelum divisualisasikan.

#### Feature Engineering Dataset Synthetic (LSTM)

```python
df_syn_fe = df_synthetic.copy()
df_syn_fe = df_syn_fe.sort_values(['user_id', 'day']).reset_index(drop=True)
print(f'Shape awal: {df_syn_fe.shape}')

# sleep_deficit harian
df_syn_fe['sleep_deficit'] = (8 - df_syn_fe['sleep_hours']).round(2)

# work_rest_ratio harian
rest_syn = (24 - df_syn_fe['work_hours_per_day'] - df_syn_fe['sleep_hours']).clip(lower=0.1)
df_syn_fe['work_rest_ratio'] = (df_syn_fe['work_hours_per_day'] / rest_syn).round(6)

# [Penanganan Outlier] cap p99 — konsisten dengan df_fe
cap_99_syn = df_syn_fe['work_rest_ratio'].quantile(0.99)
df_syn_fe['work_rest_ratio'] = df_syn_fe['work_rest_ratio'].clip(upper=cap_99_syn)
print(f'   ✅ work_rest_ratio synthetic di-cap di p99 = {cap_99_syn:.6f}')

# rolling_burnout_7d — shift(1) sebelum rolling() untuk cegah data leakage
df_syn_fe['rolling_burnout_7d'] = (
    df_syn_fe.groupby('user_id')['burnout_score']
    .transform(lambda x: x.shift(1).rolling(7, min_periods=1).mean())
    .fillna(df_syn_fe['burnout_score'])
)

# burnout_delta — perubahan burnout harian
df_syn_fe['burnout_delta'] = (
    df_syn_fe.groupby('user_id')['burnout_score']
    .transform(lambda x: x.diff().fillna(0))
)

# [Encoding Variabel ke Numerik] encoding biner harian
df_syn_fe['overwork_flag']   = (df_syn_fe['work_hours_per_day'] > THRESH_WORK_FE).astype(int)
df_syn_fe['sleep_risk_flag'] = (df_syn_fe['sleep_hours'] <= THRESH_SLEEP_FE).astype(int)
df_syn_fe['dual_risk_flag']  = ((df_syn_fe['overwork_flag'] == 1) &
                                 (df_syn_fe['sleep_risk_flag'] == 1)).astype(int)

FITUR_SYN_BARU = ['sleep_deficit', 'work_rest_ratio', 'rolling_burnout_7d',
                   'burnout_delta', 'overwork_flag', 'sleep_risk_flag', 'dual_risk_flag']

print(f'✅ {len(FITUR_SYN_BARU)} fitur baru synthetic dibuat')
print(f'   NaN di fitur baru: {df_syn_fe[FITUR_SYN_BARU].isnull().sum().sum()}')
print(f'   Shape sesudah    : {df_syn_fe.shape}')
df_syn_fe.head()
```

Tujuh fitur kontekstual harian ditambahkan ke dataset synthetic, terbagi menjadi tiga kelompok:

**Fitur numerik turunan (sama dengan burnout dataset):**
- `sleep_deficit` — defisit tidur harian vs standar WHO
- `work_rest_ratio` — rasio jam kerja terhadap jam istirahat, di-*cap* di p99 (16,51) — lebih tinggi dari burnout dataset (7,47) karena data synthetic mencakup kondisi ekstrem yang lebih luas

**Fitur kontekstual time-series (khusus LSTM):**
- `rolling_burnout_7d` — rata-rata burnout 7 hari sebelumnya per user; menggunakan `shift(1)` sebelum `rolling()` untuk mencegah *data leakage* — hari ini tidak boleh ikut menghitung rata-rata "masa lalu"-nya sendiri
- `burnout_delta` — perubahan burnout dari hari sebelumnya; nilai positif berarti memburuk, negatif berarti membaik

**Fitur biner (sama dengan burnout dataset):**
- `overwork_flag`, `sleep_risk_flag`, `dual_risk_flag`

Dari sampel output `user_0000`, fitur kontekstual bekerja dengan benar — `rolling_burnout_7d` di hari 1–2 bernilai 70 (fallback ke nilai hari itu sendiri karena belum ada histori), lalu mulai bergerak di hari 3 saat ada perubahan burnout score. `burnout_delta` = 10 di hari 3 mencerminkan kenaikan burnout dari 70 → 80 yang tepat.

> **Langkah selanjutnya:** Sliding window dibangun ulang menggunakan 7 fitur terpilih dari 15 kolom yang tersedia untuk menghasilkan input LSTM final.

#### Rebuild Sliding Window dengan Fitur Baru

```python
# [Feature Selection]
# sleep_hours DIHAPUS karena sleep_deficit = 8 - sleep_hours
# adalah transformasi linear sempurna → korelasi = -1.0 → informasi identik.
# sleep_deficit dipertahankan karena lebih interpretatif ("kurang berapa jam dari WHO").
FEATURE_COLS_FE = [
    'work_hours_per_day',   # fitur asli — jam kerja harian
    'sleep_deficit',        # FE — menggantikan sleep_hours
    'work_rest_ratio',      # FE — tekanan waktu harian (sudah di-cap p99)
    'rolling_burnout_7d',   # FE — tren 7 hari sebelumnya (no leakage)
    'burnout_delta',        # FE — arah perubahan burnout harian
    'overwork_flag',        # FE — sinyal biner kerja berlebih
    'dual_risk_flag',       # FE — sinyal kombinasi overwork + sleep risk
]

X_fe_list, y_score_fe, y_level_fe = [], [], []

for uid in df_syn_fe['user_id'].unique():
    user_df = (df_syn_fe[df_syn_fe['user_id'] == uid]
               .sort_values('day').reset_index(drop=True))
    for i in range(len(user_df) - WINDOW):
        X_fe_list.append(user_df.iloc[i:i+WINDOW][FEATURE_COLS_FE].values.tolist())
        y_score_fe.append(user_df.iloc[i+WINDOW]['burnout_score'])
        y_level_fe.append(user_df.iloc[i+WINDOW]['burnout_level'])

X_fe_arr       = np.array(X_fe_list)
y_score_fe_arr = np.array(y_score_fe)
y_level_fe_arr = np.array(y_level_fe)

print(f'✅ Rebuild sliding window selesai!')
print(f'   X shape  : {X_fe_arr.shape}  → (sampel, {WINDOW} hari, {len(FEATURE_COLS_FE)} fitur)')
print(f'   y_score  : {y_score_fe_arr.shape}')
print(f'   y_level  : {y_level_fe_arr.shape}')
print(f'\n   Fitur yang digunakan ({len(FEATURE_COLS_FE)} fitur):')
for i, f in enumerate(FEATURE_COLS_FE, 1):
    print(f'   {i}. {f}')
```

Sliding window dibangun ulang menggunakan **7 fitur terpilih** — menggantikan versi baseline 2 fitur dari Section 6. `sleep_hours` tidak dimasukkan karena `sleep_deficit = 8 - sleep_hours` adalah transformasi linear sempurna (korelasi = −1,0), sehingga keduanya membawa informasi yang identik — memasukkan keduanya sekaligus hanya menambah redudansi tanpa nilai tambah.

`sleep_risk_flag` juga tidak dimasukkan karena informasinya sudah tercakup dalam `dual_risk_flag` (`overwork_flag AND sleep_risk_flag`) — memasukkan ketiganya sekaligus berpotensi menciptakan multikolinearitas pada input LSTM.

Perbandingan sebelum dan sesudah Feature Engineering:

| | Sebelum FE | Sesudah FE |
| :--- | :--- | :--- |
| Jumlah fitur per hari | 2 | **7** |
| Shape `X_arr` | (24.495, 7, 2) | **(24.495, 7, 7)** |
| Informasi yang dibawa | Jam kerja & jam tidur mentah | + defisit, rasio, tren rolling, delta harian, sinyal risiko biner |

`X_fe_arr` ini menggantikan `X_arr` baseline sebagai input final untuk training LSTM — **array versi 2 fitur tidak perlu digunakan lagi**.

> **Langkah selanjutnya:** Scaling fitur dengan `MinMaxScaler` sebelum data siap di-export untuk training model.

#### Scaling Fitur (MinMaxScaler)

```python
# [Scaling Fitur] — MinMaxScaler sebelum LSTM
#
# Tanpa scaling, fitur dengan range besar mendominasi gradient:
#   work_hours_per_day : 4 – 19
#   rolling_burnout_7d : 10 – 100
#   burnout_delta      : -20 s/d +20
#   flag               : 0 – 1
#
# Aturan anti-leakage:
#   scaler.fit()       → HANYA pada data train
#   scaler.transform() → train, val, DAN test

n_samples  = X_fe_arr.shape[0]
n_features = X_fe_arr.shape[2]
n_train    = int(n_samples * 0.80)
n_val      = int(n_samples * 0.10)

X_train_raw = X_fe_arr[:n_train]
X_val_raw   = X_fe_arr[n_train:n_train+n_val]
X_test_raw  = X_fe_arr[n_train+n_val:]

y_score_train = y_score_fe_arr[:n_train]
y_score_val   = y_score_fe_arr[n_train:n_train+n_val]
y_score_test  = y_score_fe_arr[n_train+n_val:]

y_level_train = y_level_fe_arr[:n_train]
y_level_val   = y_level_fe_arr[n_train:n_train+n_val]
y_level_test  = y_level_fe_arr[n_train+n_val:]

# Fit HANYA pada train
scaler_fe = MinMaxScaler()
scaler_fe.fit(X_train_raw.reshape(-1, n_features))

# Transform semua split
X_train_scaled = scaler_fe.transform(
    X_train_raw.reshape(-1, n_features)).reshape(X_train_raw.shape)
X_val_scaled   = scaler_fe.transform(
    X_val_raw.reshape(-1, n_features)).reshape(X_val_raw.shape)
X_test_scaled  = scaler_fe.transform(
    X_test_raw.reshape(-1, n_features)).reshape(X_test_raw.shape)

print(f'✅ Feature Scaling (MinMaxScaler) selesai!')
print(f'   Train : {X_train_scaled.shape} | range [{X_train_scaled.min():.3f}, {X_train_scaled.max():.3f}]')
print(f'   Val   : {X_val_scaled.shape}   | range [{X_val_scaled.min():.3f},  {X_val_scaled.max():.3f}]')
print(f'   Test  : {X_test_scaled.shape}  | range [{X_test_scaled.min():.3f},  {X_test_scaled.max():.3f}]')
print(f'\n   Range asli → sesudah scaling per fitur:')
print(f'   {"Fitur":<22} {"Min":>8} {"Max":>8} {"→ Scaled":>10}')
print(f'   {"-"*50}')
for feat, mn, mx in zip(FEATURE_COLS_FE, scaler_fe.data_min_, scaler_fe.data_max_):
    print(f'   {feat:<22} {mn:>8.3f} {mx:>8.3f} {"[0.0, 1.0]":>10}')
```

Data dibagi dengan proporsi **80/10/10** (train/val/test) sebelum scaling diterapkan. Aturan anti-leakage ditegakkan ketat: `scaler.fit()` **hanya** pada data train, lalu `transform` diterapkan ke ketiga split — memastikan statistik dari val dan test tidak bocor ke proses training.

Alasan scaling diperlukan terlihat jelas dari range asli tiap fitur yang sangat berbeda:
- `rolling_burnout_7d` : 10 – 100 (range 90)
- `work_rest_ratio` : 0,235 – 190 (range sangat lebar karena outlier)
- `burnout_delta` : −10 – +10
- `overwork_flag` / `dual_risk_flag` : 0 – 1

Tanpa scaling, fitur dengan range besar seperti `rolling_burnout_7d` akan mendominasi *gradient* saat training dan membuat model mengabaikan fitur biner yang range-nya hanya 0–1. Setelah MinMaxScaler, semua fitur berada di rentang **[0,0 – 1,0]** sehingga setiap fitur berkontribusi proporsional terhadap pembelajaran model.

> ⚠️ `scaler_fe.pkl` **wajib di-load ulang saat inferensi** — data input pengguna baru harus di-transform menggunakan scaler yang sama agar skala konsisten dengan data training. Jika scaler tidak disimpan, model akan menerima input dengan skala yang berbeda dan menghasilkan prediksi yang tidak akurat.

## 9. A/B Testing


### Latar Belakang & Hipotesis

Sistem peringatan dini BurnoutLens membutuhkan keputusan desain yang kritis: **pada threshold berapa sistem harus memicu notifikasi kepada pengguna?** Threshold yang terlalu tinggi akan melewatkan banyak kasus burnout (false negative), sementara threshold terlalu rendah akan membanjiri pengguna dengan notifikasi tidak relevan (false positive).

>**Catatan Metodologi**: Eksperimen ini merupakan **simulasi perbandingan strategi threshold** (threshold comparison) menggunakan data sintetis. Kedua versi dievaluasi pada dataset yang sama — bukan pada kelompok pengguna nyata yang terpisah. Pendekatan ini dipilih karena data pengguna real belum tersedia pada tahap pengembangan produk, namun secara statistik tetap valid untuk membandingkan performa dua kebijakan sistem.

Dua strategi diuji secara head-to-head:

| | Versi A — Konservatif | Versi B — Agresif |
| :--- | :--- | :--- |
| **Threshold** | `burnout_score` ≥ **6,6** | `burnout_score` ≥ **3,6** |
| **Level yang diberi peringatan** | `High` saja | `Moderate` + `High` |
| **Filosofi** | Hanya peringatkan yang sudah kritis | Deteksi dini sebelum kritis |

**Metrik yang diukur:** Precision, Recall, F1-Score, False Positive Rate, Accuracy

**Hipotesis dirumuskan secara terpisah untuk setiap uji statistik:**

**Uji 1 — Chi-Square (Akurasi Keseluruhan)**

- **H₀**: Tidak ada perbedaan proporsi prediksi benar antara Versi A dan Versi B

- **H₁**: Terdapat perbedaan proporsi prediksi benar antara Versi A dan Versi B

**Uji 2 — Z-Test Proporsi (Recall)**

- **H₀**: Tidak ada perbedaan Recall (proporsi kasus berisiko yang terdeteksi) antara Versi A dan Versi B

- **H₁**: Terdapat perbedaan Recall antara Versi A dan Versi B

**Tingkat signifikansi**: α = 0,05 untuk kedua uji.

Recall dipilih sebagai metrik utama karena dalam konteks kesehatan mental, ***melewatkan kasus burnout (false negative) jauh lebih berbahaya** daripada mengirim notifikasi berlebih (false positive).


#### Setup & Simulasi Prediksi

```python
# ============================================================
# SECTION 9 — A/B TESTING
# ============================================================

# --- Setup & Simulasi Prediksi ---
df_ab = df_syn_fe[['user_id', 'day', 'burnout_score',
                    'burnout_level', 'dual_risk_flag',
                    'consecutive_overwork']].copy()

# Cek skala burnout_score — tentukan threshold yang benar
burnout_max = df_ab['burnout_score'].max()
print(f'Range burnout_score: {df_ab["burnout_score"].min():.1f} — {burnout_max:.1f}')

# Tentukan threshold berdasarkan skala aktual
if burnout_max > 10:
    # Skala sudah ×10 (hasil transformasi Section 6)
    THRESH_MOD_AB = 36
    THRESH_HI_AB  = 66
    print('Skala 10–100 terdeteksi → threshold: Moderate ≥ 36, High ≥ 66')
else:
    # Skala asli 1–10
    THRESH_MOD_AB = 3.6
    THRESH_HI_AB  = 6.6
    print('Skala 1–10 terdeteksi → threshold: Moderate ≥ 3.6, High ≥ 6.6')

# Ground truth: at_risk = burnout_score ≥ THRESH_MOD_AB (Moderate atau High)
df_ab['is_at_risk'] = (df_ab['burnout_score'] >= THRESH_MOD_AB).astype(int)

# Simulasi error prediksi model dengan noise Gaussian
np.random.seed(42)
noise_scale = 0.8 if burnout_max <= 10 else 8.0   # noise disesuaikan dengan skala
df_ab['predicted_score'] = (
    df_ab['burnout_score'] + np.random.normal(0, noise_scale, len(df_ab))
).clip(df_ab['burnout_score'].min(), burnout_max)

print(f'\nTotal observasi : {len(df_ab):,}')
print(f'User unik       : {df_ab["user_id"].nunique():,}')
print(f'% At Risk (GT)  : {df_ab["is_at_risk"].mean()*100:.1f}%')
display(df_ab.head(5))
```

Menyiapkan data dan mensimulasikan output prediksi model sebelum threshold diuji. Threshold disesuaikan otomatis berdasarkan skala `burnout_score` yang terdeteksi — skala 10–100 menggunakan threshold `Moderate ≥ 36` dan `High ≥ 66`.

`predicted_score` disimulasikan dengan menambahkan noise Gaussian (σ = 8,0) ke nilai aktual — merepresentasikan ketidaksempurnaan prediksi model LSTM di dunia nyata. Noise di-*clip* agar tidak melampaui batas skala.

Dari output, **70,4% observasi termasuk kategori berisiko** (`Moderate` atau `High`) — proporsi yang tinggi karena dataset synthetic didominasi anchor `High` dan `Moderate`. Ini menjadi konteks penting saat menginterpretasikan metrik A/B Testing: baseline yang tinggi berarti sistem akan sering mengirim notifikasi bahkan tanpa threshold yang agresif.

> **Langkah selanjutnya:** Dua versi threshold didefinisikan — Versi A (konservatif, hanya `High`) dan Versi B (agresif, `Moderate` + `High`) — lalu metrik keduanya dihitung dan diuji secara statistik.

####  Definisi Kedua Versi

```python
# --- Definisi Kedua Versi ---
df_ab['alert_A'] = (df_ab['predicted_score'] >= THRESH_HI_AB).astype(int)   # Konservatif
df_ab['alert_B'] = (df_ab['predicted_score'] >= THRESH_MOD_AB).astype(int)  # Agresif

for versi, col in [('A (Konservatif)', 'alert_A'), ('B (Agresif)', 'alert_B')]:
    print(f'Versi {versi}: '
          f'{df_ab[col].sum():,} peringatan ({df_ab[col].mean()*100:.1f}% observasi)')
```

Dua strategi threshold didefinisikan untuk diuji head-to-head:

| | Versi A — Konservatif | Versi B — Agresif |
| :--- | :--- | :--- |
| **Threshold** | `predicted_score` ≥ 66 | `predicted_score` ≥ 36 |
| **Level yang diberi peringatan** | `High` saja | `Moderate` + `High` |
| **Jumlah peringatan** | 12.844 (40,2% observasi) | 21.968 (68,8% observasi) |

Selisih volume peringatan antara kedua versi cukup besar — Versi B mengirim **hampir 2× lebih banyak** notifikasi dibanding Versi A. Pertanyaan utamanya: apakah peringatan tambahan tersebut tepat sasaran (mendeteksi user yang benar-benar berisiko) atau justru membanjiri pengguna dengan notifikasi yang tidak relevan?

> **Langkah selanjutnya:** Metrik Precision, Recall, F1, FPR, dan Accuracy dihitung untuk menjawab pertanyaan ini secara kuantitatif.

#### Perhitungan Metrik per Versi

```python
# --- Perhitungan Metrik per Versi ---
def hitung_metrik(y_true, y_pred, nama):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return {
        'Versi'     : nama,
        'TP': tp, 'FP': fp, 'TN': tn, 'FN': fn,
        'Precision' : round(precision_score(y_true, y_pred, zero_division=0), 4),
        'Recall'    : round(recall_score(y_true, y_pred, zero_division=0), 4),
        'F1-Score'  : round(f1_score(y_true, y_pred, zero_division=0), 4),
        'FPR'       : round(fp / (fp + tn), 4) if (fp + tn) > 0 else 0,
        'Accuracy'  : round((tp + tn) / (tp + tn + fp + fn), 4),
    }

metrik_A = hitung_metrik(df_ab['is_at_risk'], df_ab['alert_A'], 'A — Konservatif')
metrik_B = hitung_metrik(df_ab['is_at_risk'], df_ab['alert_B'], 'B — Agresif')

df_metrik = pd.DataFrame([metrik_A, metrik_B]).set_index('Versi')
display(df_metrik)
```

Fungsi `hitung_metrik` menghitung 5 metrik dari confusion matrix masing-masing versi:

| Metrik | Versi A — Konservatif | Versi B — Agresif |
| :--- | ---: | ---: |
| Precision | **1,0000** | 0,9698 |
| Recall | 0,5708 | **0,9468** |
| F1-Score | 0,7268 | **0,9582** |
| FPR | **0,0000** | 0,0702 |
| Accuracy | 0,6977 | **0,9418** |

Perbedaan yang paling mencolok ada pada **Recall** — Versi A hanya mendeteksi 57% kasus berisiko, artinya **9.658 observasi berisiko tidak mendapat peringatan sama sekali**. Versi B berhasil mendeteksi 94,7% kasus berisiko dengan *trade-off* FPR 7% — artinya dari 9.448 observasi tidak berisiko, sekitar 663 mendapat peringatan yang tidak perlu.

Dalam konteks kesehatan mental, **melewatkan kasus berisiko (False Negative) jauh lebih berbahaya** dibanding mengirim notifikasi berlebih (False Positive) yang paling buruk hanya mengganggu pengguna sejenak.

> **Langkah selanjutnya:** Uji statistik Chi-Square dan Z-Test dijalankan untuk memastikan perbedaan performa kedua versi ini signifikan secara statistik — bukan sekadar kebetulan dari noise simulasi.

#### Uji Statistik (Chi-Square & Z-Test)

```python
# --- Uji Statistik 1: Chi-Square ---
# H₀: Tidak ada perbedaan proporsi prediksi benar antara Versi A dan Versi B

# Jumlah prediksi benar dan salah per versi (dari confusion matrix)
correct_A = metrik_A['TP'] + metrik_A['TN']
wrong_A   = metrik_A['FP'] + metrik_A['FN']
correct_B = metrik_B['TP'] + metrik_B['TN']
wrong_B   = metrik_B['FP'] + metrik_B['FN']

# Tabel kontingensi tanpa smoothing — data cukup besar sehingga sel 0 tidak jadi masalah
# Jika salah satu sel = 0 (mis. Versi A FP=0), chi-square tetap valid karena total n besar
obs_table = np.array([
    [correct_A, wrong_A],
    [correct_B, wrong_B]
])

chi2, p_chi2, dof, expected = stats.chi2_contingency(obs_table)

print('=== HASIL UJI CHI-SQUARE ===')
print(f'H₀: Tidak ada perbedaan proporsi prediksi benar antara Versi A dan Versi B\n')
print(f'Tabel Observasi:')
print(f'  Versi A: {correct_A:,} benar | {wrong_A:,} salah')
print(f'  Versi B: {correct_B:,} benar | {wrong_B:,} salah')
print(f'\nChi-Square    : χ²={chi2:.4f}')
print(f'Derajat Bebas : {dof}')
print(f'p-value       : p={p_chi2:.2e}')
print(f'Alpha         : 0.05')
print(f'Keputusan     : {"✅ Tolak H₀ — proporsi prediksi benar berbeda signifikan" if p_chi2 < 0.05 else "⚠️ Gagal Tolak H₀ — tidak ada bukti perbedaan signifikan"}')
```

Uji Chi-Square membandingkan proporsi prediksi benar antara kedua versi menggunakan tabel kontingensi 2×2:

| | Prediksi Benar | Prediksi Salah |
| :--- | ---: | ---: |
| Versi A | 22.292 | 9.658 |
| Versi B | **30.090** | 1.860 |

Hasil: χ² = **6.438,68** dengan p ≈ 0,000 — **H₀ ditolak**. Perbedaan proporsi prediksi benar antara Versi A (69,8%) dan Versi B (94,2%) terbukti signifikan secara statistik, bukan kebetulan.

Nilai χ² yang sangat besar ini wajar mengingat ukuran sampel 31.950 observasi — semakin besar sampel, semakin sensitif uji Chi-Square dalam mendeteksi perbedaan. Yang terpenting adalah arah perbedaannya: Versi B menghasilkan **7.798 prediksi benar lebih banyak** dibanding Versi A.

> **Langkah selanjutnya:** Z-Test proporsi dijalankan khusus untuk membandingkan Recall kedua versi — metrik yang paling kritis dalam konteks deteksi dini kesehatan mental.

```python
# --- Uji Statistik 2: Z-Test Proporsi (Recall) ---
# H₀: Tidak ada perbedaan Recall antara Versi A dan Versi B
#
# Catatan asumsi: kedua versi dievaluasi pada populasi at-risk yang sama (n_at_risk identik).
# Ini adalah paired evaluation — bukan dua kelompok pengguna independen.
# Z-test tetap digunakan di sini untuk mengukur apakah perbedaan proporsi TP/n_at_risk
# antara kedua strategi signifikan secara statistik.

n_at_risk  = df_ab['is_at_risk'].sum()
detected_A = ((df_ab['alert_A'] == 1) & (df_ab['is_at_risk'] == 1)).sum()
detected_B = ((df_ab['alert_B'] == 1) & (df_ab['is_at_risk'] == 1)).sum()
p_A = detected_A / n_at_risk
p_B = detected_B / n_at_risk

# Two-sided: menguji apakah ada perbedaan, bukan arahnya
z_stat, p_z = proportions_ztest(
    count=[detected_A, detected_B],
    nobs=[n_at_risk, n_at_risk],
    alternative='two-sided'
)

print('=== HASIL Z-TEST PROPORSI (RECALL) ===')
print(f'H₀: Tidak ada perbedaan Recall antara Versi A dan Versi B\n')
print(f'Populasi at-risk  : {n_at_risk:,} observasi')
print(f'Recall Versi A    : {p_A:.4f} ({detected_A:,}/{n_at_risk:,} terdeteksi)')
print(f'Recall Versi B    : {p_B:.4f} ({detected_B:,}/{n_at_risk:,} terdeteksi)')
print(f'\nZ-Statistik : {z_stat:.4f}')
print(f'P-value     : {p_z:.2e}')
print(f'Alpha       : 0.05')
print(f'Keputusan   : {"✅ Tolak H₀ — Recall antara kedua versi berbeda signifikan" if p_z < 0.05 else "⚠️ Gagal Tolak H₀ — tidak ada bukti perbedaan Recall yang signifikan"}')
```

Z-Test dijalankan khusus untuk membandingkan Recall — seberapa besar proporsi kasus berisiko yang berhasil terdeteksi oleh masing-masing versi dari total 22.502 observasi berisiko.

| | Terdeteksi | Terlewat | Recall |
| :--- | ---: | ---: | ---: |
| Versi A | 12.844 | **9.658** | 0,5708 |
| Versi B | 21.305 | 1.197 | **0,9468** |

Hasil: z = −93,23 dengan p ≈ 0,000 — **H₀ ditolak**. Perbedaan Recall sebesar **0,376 poin** (57,1% vs 94,7%) terbukti signifikan secara statistik.

Angka yang paling penting di sini adalah **9.658 kasus terlewat oleh Versi A** — hampir sepertiga dari seluruh populasi berisiko tidak mendapat peringatan sama sekali. Dalam konteks aplikasi kesehatan mental, ini berarti ribuan momen intervensi dini yang terlewatkan setiap bulannya.

> **Langkah selanjutnya:** 95% Confidence Interval Recall dihitung untuk memverifikasi bahwa perbedaan ini juga signifikan secara praktis — bukan hanya statistik.

```python
# --- 95% Confidence Interval Recall ---
z_crit = stats.norm.ppf(0.975)
ci_A = (p_A - z_crit * np.sqrt(p_A*(1-p_A)/n_at_risk),
        p_A + z_crit * np.sqrt(p_A*(1-p_A)/n_at_risk))
ci_B = (p_B - z_crit * np.sqrt(p_B*(1-p_B)/n_at_risk),
        p_B + z_crit * np.sqrt(p_B*(1-p_B)/n_at_risk))

print(f'95% CI Recall A: [{ci_A[0]:.4f}, {ci_A[1]:.4f}]')
print(f'95% CI Recall B: [{ci_B[0]:.4f}, {ci_B[1]:.4f}]')

# Interpretasi CI
if ci_A[1] < ci_B[0]:
    print('\n✅ CI tidak tumpang tindih — perbedaan Recall A vs B signifikan secara praktis.')
else:
    print('\n⚠️  CI tumpang tindih — interpretasi perbedaan perlu kehati-hatian.')
```

Menghitung interval kepercayaan 95% untuk Recall masing-masing versi menggunakan metode proporsi normal:

| Versi | Recall | 95% CI |
| :--- | ---: | :--- |
| A — Konservatif | 0,5708 | [0,5643 — 0,5773] |
| B — Agresif | 0,9468 | [0,9439 — 0,9497] |

Kedua interval **tidak tumpang tindih sama sekali** — batas atas CI Versi A (0,5773) jauh di bawah batas bawah CI Versi B (0,9439), dengan jarak **0,3666 poin**. Ini membuktikan bahwa perbedaan Recall bukan sekadar signifikan secara statistik, tetapi juga **signifikan secara praktis** — tidak ada overlap yang bisa membuat kedua versi dianggap setara dalam kondisi apapun.

CI yang sangat sempit pada keduanya (lebar ~0,006) juga menunjukkan estimasi Recall yang sangat presisi — hasil dari ukuran sampel 22.502 observasi berisiko yang cukup besar.

> **Langkah selanjutnya:** Analisis tingkat user dilakukan untuk melihat berapa banyak pengguna unik yang berhasil terdeteksi oleh masing-masing versi — perspektif yang lebih relevan secara bisnis dibanding metrik per observasi.

#### Simulasi per User (User-Level Analysis)

```python
def analisis_user_level(df, alert_col):
    return df.groupby('user_id').agg(
        pernah_berisiko = ('is_at_risk', 'max'),
        terdeteksi      = (alert_col,    'max'),
    ).reset_index()

user_A      = analisis_user_level(df_ab, 'alert_A')
user_B      = analisis_user_level(df_ab, 'alert_B')
at_risk_ids = user_A[user_A['pernah_berisiko'] == 1]['user_id']

det_A  = user_A[user_A['user_id'].isin(at_risk_ids)]['terdeteksi'].sum()
det_B  = user_B[user_B['user_id'].isin(at_risk_ids)]['terdeteksi'].sum()
total  = len(at_risk_ids)

print(f'Total user pernah berisiko : {total:,}')
print(f'Versi A terdeteksi         : {det_A:,} ({det_A/total*100:.1f}%)')
print(f'Versi B terdeteksi         : {det_B:,} ({det_B/total*100:.1f}%)')
print(f'Versi A terlewat (missed)  : {total-det_A:,}')
print(f'Versi B terlewat (missed)  : {total-det_B:,}')
```

Berbeda dari metrik per observasi, analisis ini menjawab pertanyaan yang lebih relevan secara bisnis: **berapa pengguna unik berisiko yang berhasil mendapat peringatan minimal satu kali?**

| | Versi A | Versi B |
| :--- | ---: | ---: |
| User berisiko terdeteksi | 782 (85,4%) | **916 (100%)** |
| User berisiko terlewat | **134** | 0 |

Versi B berhasil mendeteksi **seluruh 916 user berisiko** tanpa satu pun yang terlewat, sementara Versi A melewatkan **134 pengguna** yang tidak pernah mendapat peringatan sama sekali sepanjang 30 hari simulasi.

Dari perspektif aplikasi kesehatan mental, 134 user yang terlewat oleh Versi A bukan sekadar angka statistik — mereka adalah pengguna yang kondisinya memburuk tanpa pernah mendapat sinyal peringatan dari sistem. Ini memperkuat rekomendasi penggunaan Versi B yang meskipun lebih agresif, memastikan tidak ada pengguna berisiko yang luput dari perhatian.

> **Langkah selanjutnya:** Seluruh hasil A/B Testing divisualisasikan dan dirangkum dalam kesimpulan akhir.
